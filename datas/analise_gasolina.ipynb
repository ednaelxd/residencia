```python
import pandas as pd
# Informando ao matplotlib sobre os gráficos em linha
%matplotlib inline
# carregar arquivo csv
# o Arquivo é separado por ponto e virgula, e codificado como cp1252
df = pd.read_csv('mensal-brasil-desde-jan2013.csv', sep=';', encoding = 'cp1252')
df.tail() # Mostra as cinco primeiras linhas da tabela
```

```python
df_limpo = df[df['PRODUTO'] == 'GASOLINA COMUM'][['MÊS','PRODUTO','PREÇO MÉDIO REVENDA','DESVIO PADRÃO REVENDA',
                                      'PREÇO MÍNIMO REVENDA','PREÇO MÁXIMO REVENDA','COEF DE VARIAÇÃO REVENDA']]
df_limpo.head()
```

```python
# Trocando vrigular por ponto
df_limpo['PREÇO MÉDIO REVENDA'] = df_limpo['PREÇO MÉDIO REVENDA'].str.replace(',', '.')
df_limpo['DESVIO PADRÃO REVENDA'] = df_limpo['DESVIO PADRÃO REVENDA'].str.replace(',', '.')
df_limpo['PREÇO MÍNIMO REVENDA'] = df_limpo['PREÇO MÍNIMO REVENDA'].str.replace(',', '.')
df_limpo['PREÇO MÁXIMO REVENDA'] = df_limpo['PREÇO MÁXIMO REVENDA'].str.replace(',', '.')
df_limpo['COEF DE VARIAÇÃO REVENDA'] = df_limpo['COEF DE VARIAÇÃO REVENDA'].str.replace(',', '.')
df_limpo.head()
```

```python
# Convertendo dados pra número
df_limpo['PREÇO MÉDIO REVENDA']= pd.to_numeric(df_limpo['PREÇO MÉDIO REVENDA'])
df_limpo['DESVIO PADRÃO REVENDA']= pd.to_numeric(df_limpo['DESVIO PADRÃO REVENDA'])
df_limpo['PREÇO MÍNIMO REVENDA']= pd.to_numeric(df_limpo['PREÇO MÍNIMO REVENDA'])
df_limpo['PREÇO MÁXIMO REVENDA']= pd.to_numeric(df_limpo['PREÇO MÁXIMO REVENDA'])
df_limpo['COEF DE VARIAÇÃO REVENDA']= pd.to_numeric(df_limpo['COEF DE VARIAÇÃO REVENDA'])
```
